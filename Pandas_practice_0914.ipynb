```python
data = {'price': [100, 200, 150, 300]}
df1 = pd.DataFrame(data)
# Apply a lambda function to calculate discounted price
df1['new_price'] = df1['price'].apply(lambda x: x * 0.9)
df1
```

```python
df1['new_price'] = df1['price']*0.9
df1
```

```python
df1['new_price'] = df1['price'].apply(lambda x: x * 0.9 if x > 150 else x*0.5)
```

```python
df1
```

```python
def cond_multiply(x):
    if x > 150:
        x = x*0.9
    else:
        x = x*0.5
    return x
df1['new_price'] = df1['price'].apply(cond_multiply)
```

```python
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df2 = pd.DataFrame(data)

condition1 = df2['A'] < 3
function1 = lambda x: x * 2
function2 = lambda x: x + 5

df2['C'] = np.where(condition1, function1(df2['B']), function2(df2['A']))

print(df2)
```
